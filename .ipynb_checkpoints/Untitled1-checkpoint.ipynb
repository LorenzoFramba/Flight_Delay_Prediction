```python
import findspark
findspark.init()

import pyspark
import random
```

```python
from pyspark import SparkContext
from pyspark.sql import SQLContext
from pyspark.sql.session import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import isnan, when, count, col
import altair as alt
```

```python
filename = "2004.csv"
```

```python
def _init_spark():
    spark = SparkSession.builder.appName("Project").getOrCreate()
    sc = spark.sparkContext
    return spark, sc
```

```python
spark, sc = _init_spark()
```

```python
sqlContext = SQLContext(sc)


df = sqlContext.read.load(filename, 
                      format='com.databricks.spark.csv', 
                      header='true',
                      delimiter=',',
                      inferSchema='true')
df.registerTempTable('dfSQL')
df.cache()
```

```python
# removing as is stated in the task along with the 'Year'
col_to_drop = ['ArrTime', 'ActualElapsedTime', 'AirTime', 'TaxiIn', 'Diverted', 
               'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay', 'Year']
df = df.drop(*col_to_drop)
```

```python
df1 = df.select('origin','dest','distance')
df2 = sqlContext.sql('select * from dfSQL where FlightNum IN (select max(FlightNum)from dfSQL)')
df3 = sqlContext.sql('select origin, dest, distance from dfSQL')
```

```python
df.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in df.columns]).show()
```

```python
df = df.filter(df.TailNum.isNotNull() )
df = df.drop('CancellationCode')
```
